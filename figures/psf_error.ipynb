```python
import os
import sys
sys.path.append('../')
import numpy as np
import json
import matplotlib.pyplot as plt
from utils.utils_plot import get_label, get_color

%matplotlib inline
```

```python
methods = {
    # 'No_Deconv':('--', 1, 4.5),
    #  'Richard-Lucy(10)':('-', 0.6, 2.5), 'Richard-Lucy(20)':('-', 0.7, 3.5),
    #  'Richard-Lucy(30)':('-', 0.8, 4.5), 'Richard-Lucy(50)':('-', 0.9, 5), 
    'Richard-Lucy(100)':('-', 1, 5),
    'FPFS':('-',1, 4.5), 
    #  'Wiener':('-', 1, 4.5),
    'Tikhonet_Laplacian':('-', 1, 4.5), 'ShapeNet':('-', 1, 4.5), 
     'Unrolled_ADMM_Gaussian(2)':('-', 0.7, 2.5),  'Unrolled_ADMM_Gaussian(4)':('-', 0.8, 3.5), 
    'Unrolled_ADMM_Gaussian(8)':('-', 1, 4.5)
}
shear_errs = [0, 0.001, 0.002, 0.003, 0.005, 0.007, 0.01, 0.02, 0.03, 0.05, 0.07, 0.1, 0.15, 0.2]
fwhm_errs = [0, 0.001, 0.002, 0.003, 0.005, 0.007, 0.01, 0.02, 0.03, 0.05, 0.07, 0.1, 0.15, 0.2]
results_path = '../results/'
```

```python

fig = plt.figure(figsize=(28,9), facecolor='white') 

# Shear Test.
gt_shear = {}
with open(os.path.join(results_path, 'No_Deconv', 'results_psf_shear_err.json'), 'r') as f:
    results = json.load(f)
for shear_err in shear_errs:
    gt_shear[str(shear_err)] = results[str(shear_err)]['gt_shear']
    
ax1 = plt.subplot(1,2,1)
for method in methods:
    results_file = os.path.join(results_path, method, 'results_psf_shear_err.json')
    with open(results_file, 'r') as f:
        results = json.load(f)
        
    g_errs, g_stds = [], []
    for shear_err in shear_errs:
        rec_shear = np.array(results[str(shear_err)]['rec_shear'])
        rec_err = np.abs(rec_shear - gt_shear[str(shear_err)])
        # rec_err = np.minimum(rec_err, np.ones_like(rec_err))
        g_errs.append(np.median(rec_err[:,2]))
        g_stds.append(np.std(rec_err[:,2])/20)
    g_errs, g_stds = np.array(g_errs), np.array(g_stds)
    
    color = get_color(method)
    if methods[method][1] == 1:
        label = get_label(method)
        ax1.plot(shear_errs, g_errs, methods[method][0], label=label, color=color, linewidth=methods[method][2], alpha=methods[method][1])
    else:
        ax1.plot(shear_errs, g_errs, methods[method][0], color=color, linewidth=methods[method][2], alpha=methods[method][1])
    
ax1.set_xlim([-0.004, 0.204])
ax1.set_xlabel('Shear Error($\Delta_{g_1}$, $\Delta_{g_2}$) in PSF', fontsize=23)
ax1.set_yscale('log')
ax1.set_ylim([0.023, 0.28])
ax1.set_yticks([0.03, 0.04, 0.05, 0.06, 0.08, 0.1, 0.15, 0.2, 0.25])
ax1.set_yticklabels(['0.03', '0.04', '0.05', '0.06', '0.08', '0.10', '0.15', '0.20', '0.25'])
ax1.set_ylabel('Median Ellipticity error \n ($\Delta g=|g_{gt} - g_{rec}|$)', fontsize=23)
ax1.legend(fontsize=20, loc='upper left')
ax1.tick_params(labelsize=17)

# FWHM Test.
gt_shear = {}
with open(os.path.join(results_path, 'No_Deconv', 'results_psf_fwhm_err.json'), 'r') as f:
    results = json.load(f)
for shear_err in shear_errs:
    gt_shear[str(shear_err)] = results[str(shear_err)]['gt_shear']

ax2 = plt.subplot(1,2,2)
for method in methods:
    results_file = os.path.join(results_path, method, 'results_psf_fwhm_err.json')
    with open(results_file, 'r') as f:
        results = json.load(f)
        
    g_errs, g_stds = [], []
    for fwhm_err in fwhm_errs:
        rec_shear = np.array(results[str(fwhm_err)]['rec_shear'])
        rec_err = np.abs(rec_shear - gt_shear[str(fwhm_err)])
        # rec_err = np.minimum(rec_err, np.ones_like(rec_err))
        g_errs.append(np.median(rec_err[:,2]))
        g_stds.append(np.std(rec_err[:,2])/20)
    g_errs, g_stds = np.array(g_errs), np.array(g_stds)

    color = get_color(method)
    if methods[method][1] == 1:
        label = get_label(method)
        ax2.plot(fwhm_errs, g_errs, methods[method][0], label=label, color=color, linewidth=methods[method][2], alpha=methods[method][1])
    else:
        ax2.plot(fwhm_errs, g_errs, methods[method][0], color=color, linewidth=methods[method][2], alpha=methods[method][1])

ax2.set_xlim([-0.004, 0.204])
ax2.set_xlabel('FWHM Error in PSF', fontsize=22)
ax2.set_yscale('log')
ax2.set_ylim([0.0238,0.09])
ax2.set_yticks([0.03, 0.04, 0.05, 0.06, 0.07, 0.08, 0.09])
ax2.set_yticklabels(['0.03', '0.04', '0.05', '0.06', '0.07', '0.08', '0.09'])
ax2.set_ylabel('Median Ellipticity Error \n ($\Delta g=|g_{gt} - g_{rec}|$)', fontsize=22)
ax2.tick_params(labelsize=17) 

plt.savefig('psf_err.jpg', bbox_inches='tight')
```
